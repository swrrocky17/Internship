# Create a Function to pass URL for Women and Men Team and get the results in DataFrame

```python
def scrap_cricket(batbowl):
    current=0
    url=batbowl
    request_object = requests.get(url)
    html_content = request_object.text
    #print(request_object.status_code, "->", url)
    soup = BeautifulSoup(request_object.content)   
    soup2 = BeautifulSoup(request_object.content)
    
    #Get the Player Names
    playerNames=[]
    for playerNameBanner in soup.find_all('div',class_="rankings-block__banner--name-large"):
        playerNames.append(playerNameBanner.text)
    for playerName in soup2.find_all('td',class_="table-body__cell rankings-table__name name"):
        playerNames.append(playerName.find('a').text.replace('\n',''))
    playerNames
    
    #Get the Nationality or Team Name
    teamNames=[]
    for teamNameBanner in soup.find_all('div',class_="rankings-block__banner--nationality"):
        teamNames.append(teamNameBanner.text.replace('\n','').replace(' ',''))
    for teamName in soup.find_all('td',class_="table-body__cell nationality-logo rankings-table__team"):
        teamNames.append(teamName.find('span',class_="table-body__logo-text").text)
    teamNames
    
    #Get Ratings 
    ratings=[]
    for ratingBanner in soup.find_all('div',class_="rankings-block__banner--rating"):
        ratings.append(ratingBanner.text)
    for rating in soup.find_all('td',class_="table-body__cell rating"):
        ratings.append(rating.text)
    ratings
    
    #Create a Dataframe
    df = pd.DataFrame({'Player Names':playerNames,'Team Name':teamNames,'Rating':ratings})
    return df.head(10)
```

# Top 10 ODI Teams, both Men and Women
Top 10 ODI teams in menâ€™s cricket along with the records for matches, points and rating.

```python
def scrap_odiTeams(odiTeam):
    current=0
    url=odiTeam
    request_object = requests.get(url)
    html_content = request_object.text
    #print(request_object.status_code, "->", url)
    soup = BeautifulSoup(request_object.content)   
    soup2 = BeautifulSoup(request_object.content)
    
    #Get the ODI Team Names
    odiTNames=[]
    for odiTNameBanner in soup.find_all('td',class_="rankings-block__banner--team-name"):
        odiTNames.append(odiTNameBanner.find('span',class_="u-hide-phablet").text)
    for odiTName in soup2.find_all('td',class_="table-body__cell rankings-table__team"):
        odiTNames.append(odiTName.find('span',class_="u-hide-phablet").text)
    odiTNames
    #len(odiTNames)
    
    #Get the no of matches played    
    #Case 1: Matches with Banner
    bannerMatches=[]    
    for matchBanner in soup.find_all('td',class_="rankings-block__banner--matches"):
        bannerMatches.append(matchBanner.text)
    bannerMatches
    
    #Case 2: Matches without Banner
    matches=[]
    for match in soup2.find_all('td',class_="table-body__cell u-center-text"):
        matches.append(match)
    #matches
    noBanner_matches=[]
    for i in range(int(len(matches))):
        if i %2 ==0:
        #print(matches[i].text)
            noBanner_matches.append(matches[i].text)
    noBanner_matches
    
    #Get the number of matches played in order by combining both lists
    total_matches = bannerMatches + noBanner_matches
    total_matches
    
    #Get Points 
    points=[]
    for p in soup.find_all('td',class_="rankings-block__banner--points"):
        points.append(p)
    for p in soup2.find_all('td',class_="table-body__cell u-center-text"):
        points.append(p)
    points
    exact_points=[]
    for i in range(int(len(points))):
        if i %2 ==0:
            exact_points.append(points[i].text)
    exact_points
    
    #Get Ratings
    ratings=[]
    for r in soup.find_all('td',class_='rankings-block__banner--rating u-text-right'):
        ratings.append(r.text.replace(' ','').replace('\n',''))
    for r in soup.find_all('td',class_='table-body__cell u-text-right rating'):
        ratings.append(r.text)
    ratings
    
    #Create a Dataframe
    df = pd.DataFrame({'Team':odiTNames,'Matches':total_matches,'Points':exact_points,'Rating':ratings})
    return df.head(10)
```

# Top 10 Men's ODI Team

```python
scrap_odiTeams("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
```

# Top 10 ODI Women Team
