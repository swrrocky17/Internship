```python
def scrap_buyNoBroker(link):
    current=0
    url=link
    request_object = requests.get(url)
    html_content = request_object.text
    #print(request_object.status_code, "->", url)
    soup = BeautifulSoup(request_object.content)
    
    #Get House Title
    titles=[]    
    for i in soup.select('h2',class_="heading-6 flex items-center font-semi-bold m-0"):
        titles.append(i.find('span',class_="overflow-hidden overflow-ellipsis whitespace-nowrap max-w-80pe po:max-w-full").get_text())
    titles
    
    #Get Locations
    locations=[]
    output_locations = []
    for i in soup.findAll('div',class_="flex"):
        if i.find('div',class_="mt-0.5p overflow-hidden overflow-ellipsis whitespace-nowrap max-w-70 text-gray-light leading-4 po:mb-0 po:max-w-95") != None:
            locations.append(i.find('div',class_="mt-0.5p overflow-hidden overflow-ellipsis whitespace-nowrap max-w-70 text-gray-light leading-4 po:mb-0 po:max-w-95").text)
        else:
            locations.append('NA')
    #locations

    for x in locations:
        if x not in output_locations:
            output_locations.append(x)
    #print(len(output_locations))
    output_locations
    
    #Get Area
    sqft=[]
    for p in soup.findAll('div',class_="flex flex-col w-33pe items-center border-r border-r-solid border-card-overview-border-color tp:w-half po:w-full last:border-r-1"):
        if p.find("div",{"id": "minRent"})!=None :
            sqft.append(p.text.replace('sqftBuiltup','sqft'))
    sqft
    
    #Get Estimated EMI
    emi=[]
    for b in soup.findAll('div',class_="flex flex-col w-33pe items-center border-r border-r-solid border-card-overview-border-color tp:w-half po:w-full last:border-r-1"):
        if b.find("div",{"id": "roomType"})!=None :
            emi.append(b.text.replace('Deposit','').replace('â\x82¹','').replace('Estimated EMI',''))
    emi
    #print(len(deposit))
    
    #Get Property Value
    propertyValue=[]
    for r in soup.findAll('div',class_="flex flex-col w-33pe items-center bo tp:w-half po:w-full border-r-0"):
        if r.find("span")!=None :
            propertyValue.append(r.text.replace('â\x82¹','').replace('Crores','Crores-').replace('Lacs','Lacs-').split('-')[0])

    propertyValue
    df = pd.DataFrame({'Property Name':titles,'Locations':output_locations,'Area':sqft,'EMI':emi,'PropertyValue':propertyValue})
    return(df)
    
```
