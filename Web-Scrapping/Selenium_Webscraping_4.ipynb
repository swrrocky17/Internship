# Q1

Scrape the details of most viewed videos on YouTube from Wikipedia.

Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos

You need to find following details:

- A) Rank
- B) Name
- C) Artist
- D) Upload date
- E) Views

```python
import selenium
from selenium import webdriver
import pandas as pd
import os
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.keys import Keys
import requests
from time import sleep
import time

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

# Import warning and ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
# Activating the chrome browser
driver=webdriver.Chrome('chromedriver.exe')
```

```python
url='https://en.wikipedia.org/'
#This opens the wikipedia.org url automatically in the Chrome and will maximise the window
driver.get(url)
driver.maximize_window()
time.sleep(2)
```

```python
driver.find_element_by_id('searchInput').send_keys('List of most-viewed YouTube videos')
```

```python
driver.find_element_by_xpath("//input[@class='searchButton']").click()
```

```python
#Create empty lists
rank=[]
name=[]
artist=[]
upload_date=[]
views=[]

# Extracting Rank from xpath
try:
    for i in driver.find_elements_by_xpath('//div[@class="mw-parser-output"]/table[3]/tbody/tr/td[1]'):
        rank.append(i.text)
except NoSuchElementException:
    rank.append("-")


# Extracting Name from xpath
try:
    for i in driver.find_elements_by_xpath('//div[@class="mw-parser-output"]/table[3]/tbody/tr/td[2]'):
        name.append(i.text)
except NoSuchElementException:
    name.append("-")
    

# Extracting Artist from xpath
try:
    for i in driver.find_elements_by_xpath('//div[@class="mw-parser-output"]/table[3]/tbody/tr/td[3]'):
        artist.append(i.text)
except NoSuchElementException:
    artist.append("-")


# Extract the Upload date from xpath
try:
    for i in driver.find_elements_by_xpath('//div[@class="mw-parser-output"]/table[3]/tbody/tr/td[4]'):
        upload_date.append(i.text)
except NoSuchElementException:
    upload_date.append("-")
    
    
# Extracting Views from xpath
try:
    for i in driver.find_elements_by_xpath('//div[@class="mw-parser-output"]/table[3]/tbody/tr/td[5]'):
        views.append(i.text)
except NoSuchElementException:
    views.append("-")
    
df=pd.DataFrame({                
                'Rank':rank,
                'Name':name,
                'Artist':artist,
                'Upload date':upload_date,
                'Availability':views
                })
df
```

```python
#CLose the driver
driver.close()
```

```python
dataset_name=[]
data_type=[]
task=[]
attribute_type=[]
instance=[]
attribute=[]
year=[]

# Extracting Dataset name from xpath
try:
    for i in driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td/table/tbody/tr/td[2]/p/b"):
        dataset_name.append(i.text)
except NoSuchElementException:
    dataset_name.append("-")
    
    
# Extracting Data type from xpath
try:
    for i in driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[2]"):
        data_type.append(i.text)
except NoSuchElementException:
    data_type.append("-")
    
# Extracting Type from xpath
try:
    for i in driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[3]"):
        task.append(i.text)
except NoSuchElementException:
    task.append("-")
    
# Extracting Attribute Type from xpath
try:
    for i in driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[4]"):
        attribute_type.append(i.text)
except NoSuchElementException:
    attribute_type.append("-")
    
# Extracting Instances from xpath
try:
    for i in driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[5]"):
        instance.append(i.text)
except NoSuchElementException:
    instance.append("-")
    
# Extracting Attribute from xpath
try:
    for i in driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[6]"):
        attribute.append(i.text)
except NoSuchElementException:
    attribute.append("-")
    
# Extracting Year from xpath
try:
    for i in driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[7]"):
        year.append(i.text)
except NoSuchElementException:
    year.append("-")
    
print (len(dataset_name))
print (len(data_type))
print (len(task))
print (len(attribute_type))
print (len(instance))
print (len(attribute))
print (len(year))
```
