# Q2 
In the above question, now scrape the following details of each product listed in first 3 pages of your 
search results and save it in a data frame and csv. In case if any product has less than 3 pages in search 
results then scrape all the products available under that product name. Details to be scraped are: "Brand 
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and 
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“

```python
brands = []
names=[]
prices=[]
rtnExchanges=[]
expDeliverys=[]
availabilities=[]
product_open_url=[]
links=[]
sleep(1)

for i in range(3):
    print('Scraping page', i+1)
    
    #Collect urls of all the laptop
    urls=driver.find_elements_by_xpath('//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    
    for i in urls:
        product_open_url.append(i.get_attribute("href"))
        sleep(1)
        
    #loop for every laptop in the list
    for url in product_open_url:
        driver.get(url)
        sleep(1)
        links.append(url)
        #exception handling for nosuchelementexception
        try: 
            #Locate the brand and then append
            brand=driver.find_element_by_xpath('//span[@class="a-size-large product-title-word-break"]')
            brands.append(brand.text.split(' ')[0]) #appending the brands in Brand list        
        except NoSuchElementException:
            brands.append("-")
            
        try:
            #Locate the name and then append
            name=driver.find_element_by_xpath('//span[@class="a-size-large product-title-word-break"]')
            names.append(name.text) #appending the names in name list
        except NoSuchElementException:
            names.append("-")
            
        try:
            #Locate the price of each product
            price=driver.find_element_by_xpath('//span[@class="a-price aok-align-center reinventPricePriceToPayPadding priceToPay"]//span[@class="a-offscreen"]')
            prices.append(price.text) #append the prices to the price List
        except NoSuchElementException:
            prices.append("-")
        
        try:
            #Locate the Return Policy of each product
            rtnExchange=driver.find_element_by_xpath('//div[@id="RETURNS_POLICY"]//a')
            rtnExchanges.append(rtnExchange.text) #append the Return policy to the Retrun Exchange List
        except NoSuchElementException:
            rtnExchanges.append("-")
            
        #expDeliverys
        try:
            #Locate the Expected Delivery of each product
            expDelivery=driver.find_element_by_xpath('//div[@id="mir-layout-DELIVERY_BLOCK-slot-DELIVERY_MESSAGE"]//b')
            expDeliverys.append(expDelivery.text) #append the Expected Delivery to the Expected Delivery List
        except NoSuchElementException:
            expDeliverys.append("-")
            
        #availabilities
        try:
            #Locate the availability of each product
            availability=driver.find_element_by_xpath('//span[@class="a-size-medium a-color-success"]')
            availabilities.append(availability.text) #append the availability to the availability List
        except NoSuchElementException:
            availabilities.append("-")       
        
            
            
df=pd.DataFrame({                
                'Brand Name':brands,
                'Name of the Product':names,
                'Price':prices,
                'Return/Exchange':rtnExchanges,
                'Expected Delivery':expDeliverys,
                'Availability':availabilities,    
                'Product URL':links,
                })
#Print the dataframe
df

#df.to_csv(r"C:\Users\mum1user\Internship\Webscrapping\Selenium\urls.csv")
       
```

# Q7. 
Write a program to scrap all the available details of best gaming laptops from digit.in

```python
def digit_in():
    driver=webdriver.Chrome('chromedriver.exe')
    url='https://www.digit.in/'
    #This opens the digit.in url automatically in the Chrome and will maximise the window
    driver.get(url)
    driver.maximize_window()
    time.sleep(2)
    
    #Finding web element for search job by using class name
    search_job=driver.find_element_by_xpath('//img[@alt="Search button"]')
    search_job.click()

    input_=driver.find_element_by_xpath('//input[@id="globalPageSearchText"]')
    input_.send_keys('Gaming Laptop')
    input_.send_keys(Keys.ENTER)
    
    #Now we have to select Best Gaming Laptops, so lets Select Top10 radio button from the site
    top_10=driver.find_element_by_xpath('//input[@id="content_top10"]')
    top_10.click()
    
    #From here Select Best Gaming Laptop in India
    best_laptop=driver.find_element_by_xpath('/html/body/div[6]/div[2]/div[2]/div[7]/a/div')
    best_laptop.click()
    
    #Best Gaming Laptop Site is opened, now lets scrap all the details now
    
    select_name=driver.find_elements_by_tag_name('h3')
    names=[]
    for i in select_name:
        names.append(i.text) #append the name
    names
    
    select_OS=driver.find_elements_by_xpath('//div[@class="Spcs-details"]//tr[3]//td[3]')
    os=[]
    for i in select_OS:
        os.append(i.text) #append the OS
    os

    select_Display=driver.find_elements_by_xpath('//div[@class="Spcs-details"]//tr[4]//td[3]')
    display=[]
    for i in select_Display:
        display.append(i.text) #append the display
    display

    select_Processor=driver.find_elements_by_xpath('//div[@class="Spcs-details"]//tr[5]//td[3]')
    processor=[]
    for i in select_Processor:
        processor.append(i.text) #append the processor
    processor

    select_memory=driver.find_elements_by_xpath('//div[@class="Spcs-details"]//tr[6]//td[3]')
    memory=[]
    for i in select_memory:
        memory.append(i.text) #append the memory
    memory

    select_weight=driver.find_elements_by_xpath('//div[@class="Spcs-details"]//tr[7]//td[3]')
    weight=[]
    for i in select_weight:
        weight.append(i.text) #append the weight
    weight

    select_dimenstion=driver.find_elements_by_xpath('//div[@class="Spcs-details"]//tr[8]//td[3]')
    dimension=[]
    for i in select_dimenstion:
        dimension.append(i.text) #append the dimension
    dimension

    graphics_Processor=driver.find_elements_by_xpath('//div[@class="Spcs-details"]//tr[9]//td[3]')
    graphics=[]
    for i in graphics_Processor:
        graphics.append(i.text) #append the Graphics
    graphics

    #Creat the Dataframe
    df=pd.DataFrame({'Name':names,
                     'OS':os,
                     'Display':display,
                     'Processor':processor,
                     'Memory':memory,
                     'Weight':weight,
                     'Dimension':dimension,
                     'Graphics Processor':graphics
                    })
    return(df)
    driver.close()
```

# Q8. 
Write a python program to scrape the details for all billionaires from www.forbes.com. Details to be 
scrapped: “Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”.


```python
def forbes():
    driver=webdriver.Chrome('chromedriver.exe')
    url='https://www.forbes.com/billionaires/'
    #This opens the forbes.com url automatically in the Chrome and will maximise the window
    driver.get(url)
    driver.maximize_window()
    time.sleep(2)

    ranks=driver.find_elements_by_xpath('//div[@class="rank"]')
    rank=[]
    for i in ranks:
        try:
            rank.append(i.text)
        except NoSuchElementException:
            rank.append("-")
    rank

    name=driver.find_elements_by_xpath('//div[@class="personName"]')
    name_forbes=[]
    for i in name:
        try:
            name_forbes.append(i.text)
        except NoSuchElementException:
            name_forbes.append("-")
    name_forbes

    net_worths=driver.find_elements_by_xpath('//div[@class="netWorth"]')
    net_worth=[]
    for i in net_worths:
        try:
            net_worth.append(i.text)
        except NoSuchElementException:
            net_worth.append("-")
    net_worth

    ages=driver.find_elements_by_xpath('//div[@class="age"]')
    age=[]
    for i in ages:
        try:
            age.append(i.text)
        except NoSuchElementException:
            age.append("-")
    age

    citizenships=driver.find_elements_by_xpath('//div[@class="countryOfCitizenship"]')
    citizenship=[]
    for i in citizenships:
        try:
            citizenship.append(i.text)
        except NoSuchElementException:
            citizenship.append("-")
    citizenship

    sources=driver.find_elements_by_xpath('//div[@class="source-column"]')
    source=[]
    for i in sources:
        try:
            source.append(i.text)
        except NoSuchElementException:
            source.append("-")
    source

    industries=driver.find_elements_by_xpath('//div[@class="category"]')
    industry=[]
    for i in industries:
        try:
            industry.append(i.text)
        except NoSuchElementException:
            industry.append("-")
    industry
    df=pd.DataFrame({'Rank':rank,
                'Name':name_forbes,
                'Net Worth':net_worth,
                'Age':age,
                'Citizenship':citizenship,
                'Source':source,
                'Industry':industry
                })
    return(df)
    driver.close()
```

```python
sleep(4)
hostel_nameLists=driver.find_elements_by_xpath('//h2[@class="title title-6"]')
hostel_nameList=[]
for i in hostel_nameLists:
    try:
        hostel_nameList.append(i.text) #append Hostelname in Hostelname List 
    except NoSuchElementException:
        hostel_nameList.append("-")
hostel_nameList
```

```python
distance_FromCities=driver.find_elements_by_xpath('//a[@class="show-on-map"]//span[@class="description"]')
distance_FromCity=[]
for i in distance_FromCities:
    try:
        distance_FromCity.append(i.text) #append Distance from city on the DistanceFromCity List
    except NoSuchElementException:
        distance_FromCity.append("-")
distance_FromCity
```

```python
ratings=driver.find_elements_by_xpath('//div[@class="score orange big"]')
rating=[]
for i in ratings:
    try:
        rating.append(i.text) #append rating in Ratings Lits
    except NoSuchElementException:
        rating.append("-")
rating
```

```python
totalReviews=driver.find_elements_by_xpath('//div[@class="reviews"]')
totalReview=[]
for i in totalReviews:
    try:
        totalReview.append(i.text) #append TotalReview in TotalReview List
    except NoSuchElementException:
        totalReview.append("-")
totalReview
```

```python
overallReviews=driver.find_elements_by_xpath('//*[@id="__layout"]//div[2]/div[2]/a/div/div[2]/div[1]/span')
overallReview=[]
for i in overallReviews:
    try:
        overallReview.append(i.text) #append OverallReviews in the OverallReview List
    except NoSuchElementException:
        overallReview.append("-")
overallReview
```

```python
facilities=driver.find_elements_by_xpath('//div[@class="facilities-label facilities"]')
facility=[]
for i in facilities:
    try:
        facility.append(i.text.replace('\n',','))
    except NoSuchElementException:
        facility.append("-")
facility
```

```python
#Collect urls of all the Hotels
product_open_url=[]
description=[]
urls=driver.find_elements_by_xpath('//h2[@class="title title-6"]//a')    
for i in urls:
    product_open_url.append(i.get_attribute("href"))
len(product_open_url)
```

```python
private=[]
for url in product_open_url:
    driver.get(url)
    sleep(1)
    try:
        try:            
            privates=driver.find_element_by_xpath('//*[@id="availability-section"]/div/section/div/div[1]/div[1]/div[2]/div[1]/div[2]/div[1]/div/div[1]/div[1]')        
            private.append(privates.text) #appending the PrivateRoom prices in PrivateRoom Price list 
            sleep(1)
        except:
            privates=driver.find_element_by_xpath('//*[@id="availability-section"]/div/section/div/div[1]/div[1]/div[2]/div[1]/div[2]/div[2]/div/div[1]/div[2]')
            private.append(privates.text) #appending the PrivateRoom prices in PrivateRoom Price list                                              
    except NoSuchElementException:
        private.append('-')
private
```

```python
dorm=[]
for url in product_open_url:
    driver.get(url)
    sleep(1)
    try:  
        try:
            dorms=driver.find_element_by_xpath('//div/div[1]/div[2]/div[2]/div[1]/div[2]/div[1]/div/div[1]/div[1]/div')        
            dorm.append(dorms.text) #appending the Dorms in Dorms list 
            sleep(1)      
        except:
            dorms=driver.find_element_by_xpath('//div[2]/div[2]/div[1]/div[2]/div[1]/div/div[1]/div[2]/div[1]')
            dorm.append(dorms.text) #appending the Dorms in Dorms list 
    except NoSuchElementException:
        dorm.append('-')
dorm
```

```python
for url in product_open_url:
    driver.get(url)
    try:
        descriptions=driver.find_element_by_xpath('//*[@id="__layout"]//div[6]/div/div[2]/div[2]/div/div[2]')        
        description.append(descriptions.text) #appending the descriptions in descriptions list        
    except NoSuchElementException:
        descriptions=driver.find_element_by_xpath('//*[@id="__layout"]//div[6]/div/div[2]/div/div/div[2]')
        description.append(descriptions.text)
len(description)
```
